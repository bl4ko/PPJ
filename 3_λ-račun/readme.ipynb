
### λ-račun

Sintaksa izrazov:

- **spremenljivke**: x, y, z, ...
- **abstrakcija**: λx.e (x se slika v e)
    - λ-abstrakcija veze kolikor se da: `λx.e1 e2 e3= λx.(e1 e2 e3)`
    - e1 (λx . e2 e3 e4) e5 = (e1 ( λx. (e2 e3 e4) )) e5
    - λxy . f(λz.zz)(λ.x) = λx . (λy . (f(λz . zz) (λx . x)))
- **uporaba ali aplikacija**: e1 e2 (funkcijo e1 uporabi na e2)
    - levo asociativna: `e1 e2 e3 = (e1 e2) e3`


Računska pravila:

- β-redukcija `(λx.e)e' = e[e'/x]`
    - konfluenca, vrstni red redukcije ni pomemben (vsi načini računanja vodijo do istega rezultat)


### Strategiji računanja

(λx.e1) ((λy.e2) e3)

1. Možnost
    - (λx.e1) ((λy.e2) e3) = (λx.e1) (e2[e3/y]) = e1[e2[e3/y]/x] (**Neučakana strategija**)
2. Možnost
    - (λx.e1) ((λy.e2) e3) = (λy.e2) e3[e1/x] = e2[e3/e1] (**Lena strategija**)

### Programiranje v λ-računu

**identiteta**: `λx.x`

**Pogojni stavek**:

- `if := λbuv. buv`
    - `if true A B = (λ b u v . b u v) true A B = (λ u v . true u v) A B = (λv. true v) B = true B = (λxy.x) B = B`



### Boolova števila

```
true := \ x y . x ;
false := \ x y . y ;
not := \ x . x false true ;
and := \ x y . x y false ;
or := \ x y . x true y ;
```

### Števila

Churchevi numerali: 

- `0 := λf.λx.x`
- `1 := λf.λx.f(x)`
- `2 := λf.λx.f(f(x))`
- `3 := λf.λx.f(f(f(x))))`
- `4 := λf.λx.f(f(f(f(x))))`

`n g a` g uporabi n-krat na a

Naslednik:

- `succ := λn.λf.λx.f(n f x)`
- `plus := λm.λn.λf.λx.m f (n f x)`
- `times := λm.λn.λf.λx.m (n f) x`


Primer λ calculator:
```
:eager
:deep
zero := \ f x . x ;
one := \ f x . f x ;
suc := \ n f x . f (n f x) ;
plus := \ n m f x . n f (m f x) ;
times := \ n m f x . n (m f) x ;

two := suc one ;
three := suc (suc (suc zero)) ;
```
